```python
#removimg the null values
house.dropna(inplace= True)
#removing the duplicate values
house.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
house.groupby("condition of the house")["Price"].mean().sort_values(ascending=True).plot(kind='bar')
plt.title("Condition of the house vs prices")
plt.ylabel("Mean Prices")
plt.xlabel("Condition of the house")
plt.show()
```

```python
x =house[["number of bedrooms","number of bathrooms","living area","condition of the house",'Number of schools nearby']]
y =house[["Price"]]
```

```python
x
```

```python
y
```

```python
house.shape[0] * 0.8
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    "criterion": ["mse","friedman_mse","mae"],
    "splitter": ["best","random"],
    "max_depth" : [None,10,20,30,40,50],
    "min_samples_split" : [2,5,10],
    "min_samples_leaf" : [1,2,4]
}
```

```python
tree_model =DecisionTreeRegressor()
```

```python
grid_tree= GridSearchCV(estimator =tree_model,param_grid=param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
grid_tree.best_params_
```

```python
tree_pred =grid_tree.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,tree_pred)
```

```python
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
```

```python
Lr.fit(x_train,y_train)
```

```python
predlr= Lr.predict(x_test)
```

```python
mean_absolute_error(y_test,predlr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfmodel= RandomForestRegressor()
```

```python
param_gridrf={
    "max_depth":[5,10,15],
    "n_estimators":[2,3,4,5,6,7,8,9,10]
}
```

```python
gridrfr = GridSearchCV(rfmodel,param_gridrf)
```

```python
gridrfr.fit(x_train,y_train)
```

```python
gridrfr.fit(x_train,y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfpred= gridrfr.predict(x_test)
```

```python
mean_absolute_error(y_test,rfpred)
```
